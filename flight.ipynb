```python
df=df.drop(['Unnamed: 0', 'flight'], axis=1)
```

```python
df
```

```python
dfe = pd.get_dummies(df[['airline', 'source_city', 'destination_city', 'departure_time', 'arrival_time', 'stops','class']], drop_first=True, dtype=int)
dfe
```

```python
dfe.dtypes
```

```python
df_1=pd.concat([df,dfe],axis=1)
df_1
```

```python
df_1.drop(['airline','source_city','departure_time','stops','arrival_time','destination_city','class'],axis=1,inplace=True)
df_1
```

```python
x=df_1.drop(['price'],axis=1)
x
```

```python
y=df_1['price']
y
```

```python
#Traininig and testing the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30,random_state=42)
x_train
x_test
```

```python
from sklearn.linear_model import LinearRegression

#creating object of Linear Regression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#data frame
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1
```

```python
#difference between actual and predicted values
df1['Difference'] = df1['Actual'] - df1['Predicted']
df1['Difference']
```

```python
#performance evaluation methods
#mean absolute error
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
```

```python
#absolute percentage error
from sklearn import metrics
print('mean absolute percentage error:', metrics.mean_absolute_percentage_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
#R2 score
from sklearn.metrics import r2_score
print('R2 Score:', r2_score(y_test, y_pred))
```
